# New Section

```
def left_diamond(n):
  for i in range(n):
    for j in range(i+1):
      print("*", end = " ")
    print('\n')
  for k in range(n):
    for l in range(k,n-1):
      print("*", end = " ")
    print('\n')


n = int(input("Enter a no. "))
left_diamond(n)
```

```
def one_zero_pattern(n):
  for i in range(n):
    for j in range(i+1):
      if (i+j)%2==0:
        print('1', end = " ")
      else:
        print('0', end = " ")
    print('\n')

n= int(input("Enter a no. "))
one_zero_pattern(n)
```

```
def numbered_gate(n):
  for i in range(n):
    for j in range(1,i+2):
      print(j, end = " ")
    for k in range(2*n-2*i-2):
      print(' ', end = " ")
    for l in range(i+1,0,-1):
      print(l, end = " ")
    print('\n')

n = int(input("Enter a no. "))
numbered_gate(n)

```

```
def countng(n):
  for i in range(n):
    for j in range(j+1,i+2):
      print(j, end = " ")
    print('\n')

n = int(input("Enter a no. "))
countng(n)
```

```
def number_counter(n,c=0):
  for i in range(n):
    for j in range(1,i+2):
      c += 1
      print(c, end = " ")
    print('\n')

n = int(input("Enter a number. "))
number_counter(n)
```

```
def print_alphabet(n):
  for i in range(n):
    for j in range(65,65+i+1):
      print(chr(j), end = " ")
    print('\n')

n = int(input("Enter a no. "))
print_alphabet(n)
```
