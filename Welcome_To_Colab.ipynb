```
def left_diamond(n):
  for i in range(n):
    for j in range(i+1):
      print("*", end = " ")
    print('\n')
  for k in range(n):
    for l in range(k,n-1):
      print("*", end = " ")
    print('\n')


n = int(input("Enter a no. "))
left_diamond(n)
```

```
def one_zero_pattern(n):
  for i in range(n):
    for j in range(i+1):
      if (i+j)%2==0:
        print('1', end = " ")
      else:
        print('0', end = " ")
    print('\n')

n= int(input("Enter a no. "))
one_zero_pattern(n)
```

```
def numbered_gate(n):
  for i in range(n):
    for j in range(1,i+2):
      print(j, end = " ")
    for k in range(2*n-2*i-2):
      print(' ', end = " ")
    for l in range(i+1,0,-1):
      print(l, end = " ")
    print('\n')

n = int(input("Enter a no. "))
numbered_gate(n)

```

```
def countng(n):
  for i in range(n):
    for j in range(j+1,i+2):
      print(j, end = " ")
    print('\n')

n = int(input("Enter a no. "))
countng(n)
```

```
def number_counter(n,c=0):
  for i in range(n):
    for j in range(1,i+2):
      c += 1
      print(c, end = " ")
    print('\n')

n = int(input("Enter a number. "))
number_counter(n)
```

```
def print_alphabet(n):
  for i in range(n):
    for j in range(65,65+i+1):
      print(chr(j), end = " ")
    print('\n')

n = int(input("Enter a no. "))
print_alphabet(n)
```

```
def inverted_alphabet(n):
    for i in range(n):
        for j in range (65, 65+n-i):
            print(chr(j), end = " ")
        print('\n')

n = int(input("Enter a no. "))
inverted_alphabet(n)
```

```
def left_trianle_alphabet(n):
  for i in range(n):
    for j in range(65+i,65+i+1):
      print(chr(j)*(i+1), end = " ")
    print('\n')

n = int(input("Enter a no. "))
left_trianle_alphabet(n)
```

```
def middle_alpha_triangle(n):
  for i in range(n):
    for j in range(n-i-1):
      print(" ", end = " ")
    for k in range(65, 65+i+1):
      print(chr(k), end = " ")
    for l in range(64+i, 64,-1):
      print(chr(l), end = " ")
    print('\n')

n = int(input("Enter a no. "))
middle_alpha_triangle(n)
```

```
def inverted_alpha_left(n):
  for i in range(n):
    for j in range(65+n-i-1, 65+n):
      print(chr(j), end = " ")
    print('\n')

n = int(input("Enter a no. "))
inverted_alpha_left(n)
```

```
def empty_box(n):
  for i in range(n):
    for j in range(n-i):
      print("*", end = " ")
    for k in range(2*i):
      print(" ", end = " ")
    for l in range(n-i):
      print("*", end = " ")
    print('\n')

n = int(input("Enter a no. "))
empty_box(n)
```
