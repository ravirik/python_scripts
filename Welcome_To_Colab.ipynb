# New Section

```
def left_diamond(n):
  for i in range(n):
    for j in range(i+1):
      print("*", end = " ")
    print('\n')
  for k in range(n):
    for l in range(k,n-1):
      print("*", end = " ")
    print('\n')


n = int(input("Enter a no. "))
left_diamond(n)
```
